```python
# Question 3.2
tokens = []
count_1 = 0
count_2 = 0

# tokenizing the data set
for word in corpus:
    tokens.append(nltk.word_tokenize(word))

# Splitting into training and testing sets
X_emo_train, X_emo_test, y_emo_train, y_emo_test = train_test_split(tokens, y_emo, test_size=0.2, random_state=0)
X_sent_train, X_sent_test, y_sent_train, y_sent_test = train_test_split(tokens, y_sent, test_size=0.2, random_state=0)

# Counting tokens for training sets
for token in X_emo_train:
    count_1 = count_1 + len(token)
    
for token in X_sent_train:
    count_2 = count_2 + len(token)
    
print("Number of tokens for the emotions training set:", count_1)
print("Number of tokens for the sentiments training set:", count_2)
```

```python
# Function to calculate the average embedding of posts and hit rates
def embeddedAvg (X, y, wv, select):
    X_average = []
    y_average = []
    count_hit_rates = 0
    count_missed_rates = 0

    for j in range(len(X)):
        count_average = 0
        vec_sum = 0
    
        for i in range(len(X[j])):
            try:
                vec_sum = wv[X[j][i]] + vec_sum
                count_hit_rates = count_hit_rates + 1
                count_average = count_average + 1 
            except KeyError:
                count_missed_rates = count_missed_rates + 1
            
        if count_average != 0:
            X_average.append(vec_sum / count_average)
            y_average.append(y[j])

    match select:
        case 1:
            print(f"Overall hit rate of emotions training set = {count_hit_rates/(count_missed_rates + count_hit_rates)*100} %\n")
        
        case 2:
            print(f"Overall hit rate of emotions testing set = {count_hit_rates/(count_missed_rates + count_hit_rates)*100} %\n")
        
        case 3:
            print(f"Overall hit rate of sentiments training set = {count_hit_rates/(count_missed_rates + count_hit_rates)*100} %\n")
        
        case 4:
            print(f"Overall hit rate of sentiments testing set = {count_hit_rates/(count_missed_rates + count_hit_rates)*100} %\n")
    
    return X_average, y_average
```

```python
# Question 3.3 & 3.4 - Embedding averages of posts and hit rates
results = embeddedAvg(X_emo_train, y_emo_train, wv, 1)
X_average_emo_train = results[0]
y_average_emo_train = results[1]

results = embeddedAvg(X_emo_test, y_emo_test, wv, 2)
X_average_emo_test = results[0]
y_average_emo_test = results[1]

results = embeddedAvg(X_sent_train, y_sent_train, wv, 3)
X_average_sent_train = results[0]
y_average_sent_train = results[1]

results = embeddedAvg(X_sent_test, y_sent_test, wv, 4)
X_average_sent_test = results[0]
y_average_sent_test = results[1]
```
